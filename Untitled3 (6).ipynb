## 과제8 : Hashing 문제 풀이

**1번**


**답:**

*   장점

1.   개방주소법 : 추가적인 메모리 할당 없이 테이블 내에서 해결 가능 (공간 효율성)
2.   체이닝 : 테이블이 꽉 차지 않으며, 삭제 연산이 상대적으로 간편함


*   단점

1.   개방주소법 : 데이터가 특정 구역에 몰리는 군집화 현상 발생 가능
2.   체이닝 : 리스트를 위한 추가 메모리 필요, 리스트가 길어지면 검색 성능 저하

---

**2번 (1)**

**답:** 테이블 크기는 1,204인데 해시 함수는 1,024를 나머지의 기준으로 사용하고 있다. 해시 결과값은 항상 0 ~ 1023 범위에만 머물게 된다. 따라서 인덱스 1024부터 1203까지의 공간(180개 버킷)은 절대로 사용되지 않아 메모리가 낭비된다.

---

**2번 (2)**

**답:** 해시 함수는 동일한 입력값 x에 대해 항상 동일한 결과값을 반환해야 하는데, 위 함수는 실행할 때마다 random 값이 변하므로, 데이터를 저장할 때의 위치와 나중에 찾을 때의 위치가 달라져 검색이 불가능해진다.

**7번**

**답:** 리스트의 맨 앞에 삽입하면 $O(1)$의 시간이 걸리지만, 맨 뒤에 삽입하려면 리스트의 끝까지 찾아가야 하므로 삽입 연산의 시간 복잡도가 $O(n)$으로 늘어난다.

---

**8번**

**답:** 삽입 : 정렬된 위치를 찾아야 하므로 기존 $O(1)$에서 리스트를 탐색하는 $O(n)$으로 효율성이 나빠진다.

삭제 : 먼저 삭제할 대상을 검색해야 하므로 검색 효율의 변화에 따른다.

성공적인 검색 : 큰 변화가 없다. (어차피 찾는 값이 있을 때까지 가야함)

실패하는 검색 : 리스트 전체를 볼 필요 없이, 찾는 값보다 큰 값이 나오면 즉시 멈출 수 있으므로 평균적인 효율성이 좋아진다.
